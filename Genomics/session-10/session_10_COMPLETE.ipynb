## Exercise 1: Forward Diffusion

```python
def forward_diffusion(x0, t, noise_schedule):    noise = torch.randn_like(x0)    alpha_bar = noise_schedule[t]    xt = torch.sqrt(alpha_bar) * x0 + torch.sqrt(1 - alpha_bar) * noise    return xt, noise# Beta schedulebetas = torch.linspace(0.0001, 0.02, 1000)alphas = 1 - betasalpha_bar = torch.cumprod(alphas, dim=0)
```
